```python
df_produtos = pd.read_csv('dados/produtos.csv', delimiter=',')
df_produtos
```

```python
df_produtos.dtypes
```

```python
df_produtos['Quantidade'] = pd.to_numeric(df_produtos['Quantidade']) # mudando o tipo da coluna para numérico (está como int 64)
df_produtos['Preço'] = pd.to_numeric(df_produtos['Preço']) # mudando o tipo da coluna para número (está como int64)
df_produtos.dtypes
```

```python
# 1 - Desempenho de vendas por categoria
vendas_categoria = df_produtos.groupby('Categoria')['Quantidade'].sum()
vendas_categoria
```

```python
# 2 - produto mais vendido
mais_vendido = df_produtos.groupby('Produto')['Quantidade'].sum().idxmax()
mais_vendido
```

```python
# 3 - Total de unidades vendidas por produto
unidades_vendidas_por_produto = df_produtos.groupby('Produto')['Quantidade'].sum()
unidades_vendidas_por_produto
```

```python
# 4 - Preço médio por produto
media_preco = df_produtos.groupby('Produto')['Preço'].mean()
media_preco
```

```python
# 5 - Correlação entre unidades vendidas e preço unitário - se o preço está surtindo efeito na quantidade de vendas
correlacao = df_produtos['Quantidade'].corr(df_produtos['Preço'])
correlacao
# Resposta: deu negativo: np.float64(-0.7228821786680254), quanto maior o preço, menor a quantidade vendidade. Deu negativo, quer dizer que a correção está ruim
```

```python
# Usando a planilha para criar uma planilha específica com outras duas colunas, Categoria e UnidadesVendidas
vendas_categoria_df = vendas_categoria.reset_index()
vendas_categoria_df.columns = ['Categoria', 'UnidadesVendidas']
vendas_categoria_df
```
