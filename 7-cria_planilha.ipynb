```python
import pandas as pd
```

```python
dados = {
    'Times': ['Flamengo', 'Palmeiras', 'São Paulo', 'Santos', 'Corinthians'],
    'Partidas_jogadas': [10, 10, 10, 10, 10],
    'Vitorias': [7, 5, 6, 4, 3],
    'Empates': [2, 3, 1, 3, 2],
    'Derrotas': [1, 2, 3, 3, 5],
    'Gols_pro': [22, 18, 16, 14, 10],
    'Gols_contra': [10, 12, 15,16, 20]


}
```

```python
df = pd.DataFrame(dados)
df
```

```python
df['Saldo_gols'] = df['Gols_pro'] - df['Gols_contra'] # Criando uma nova coluna
df
```

```python
df['Pontos'] = df['Vitorias'] * 3 + df['Empates'] * 1 # Criand uma nova coluna de pontos
df
```

```python
df = df.sort_values(by='Pontos', ascending=False) # Mostrando a classificação de maior número de pontos para menor número de pontos
df
```
